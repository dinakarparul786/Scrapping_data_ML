```python
from bs4 import BeautifulSoup

with open("scrapped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
all_h3 = soup.find_all("h3")

all_countries = []

for h3 in all_h3:
    country_name = h3.get_text(strip=True)

    country_population = h3.find_next("div").select_one("span.country-population").get_text(strip=True)
    #another way 
    #print(h3.find_text("div").select("span.country-population")[0].get_text(strip=True))
    #try .children .next_sibilings .previous_sibilings .parent functions 
    #and to access atrributes in a list we can use indexing ex: h3.find_text("div").select("span.country-population")[0]

    all_countries.append([country_name,country_population])

print(all_countries)
    
```

```python
import pandas as pd

df = pd.DataFrame(all_countries,columns=["Name","Population"])

df.to_csv("cleaned_data/data.csv",index=False)
```
