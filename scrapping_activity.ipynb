```python
import requests
from bs4 import BeautifulSoup
```

```python
#URL = "https://quotes.toscrape.com/page/1/"
page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text,"lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("There is no valid quote page")
        break

    with open(f"scrapped_data/quote{page_count}.html","w",encoding="utf-8") as f:
        f.write(res.text)
        print(f"downloaded data from quote{page_count}")
        

    page_count = page_count + 1
```
